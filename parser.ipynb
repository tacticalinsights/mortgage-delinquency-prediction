```python
delinquent = set()
for yr in range(1999, 2022):
    for q in range(1, 5):
        with open(home + f"\\historical_data_{yr}Q{q}\\historical_data_time_{yr}Q{q}.txt") as f:
            content = f.readlines()
            for line in content:
                if line.split('|')[3] == '1':
                    delinquent.add(line.split('|')[0])
delinquent = np.array(list(delinquent))
```

```python
i = 0
for yr in range(2019, 2022):
    for q in range(1, 5):
        with open(home + f"\\historical_data_{yr}Q{q}\\historical_data_{yr}Q{q}.txt") as f:
            content = f.readlines()
            for j in range(31):
                lst = []
                for line in content:
                    lst.append(line.split('|')[j])
                datasets[i][cols[j]] = lst
        datasets[i].set_index('Loan Sequence Number', inplace = True)
        datasets[i]['Delinquent'] = datasets[i].index.isin(delinquent).astype(int)
        datasets[i].to_csv(home + f"\\Data\\{yr}_{q}" + '.csv')
        i += 1
df = pd.concat(datasets)
df.to_csv(home + f"\\full_data.csv")
```
